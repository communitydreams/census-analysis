```python
import pandas as pd
import json
```

```python
df = pd.read_excel('data/all-geocodes-v2022.xlsx', skiprows=3, header=1, dtype=str)
df
```

```python
state_name_mapping = df.loc[
    (df['County FIPS Code'] == '000') & 
    (df['County Subdivision FIPS Code'] == '00000') & 
    (df['Place FIPS Code'] == '00000') & 
    (df['Consolidated City FIPS Code'] == '00000'),
    ['State FIPS Code', 'Area Name']
].set_index('State FIPS Code')['Area Name'].to_dict()
state_name_mapping
```

```python
df['State Name'] = df['State FIPS Code'].map(state_name_mapping)
df
```

```python
county_df = df[df['County FIPS Code'] != '000']
county_df
```

```python
county_df = county_df[['State FIPS Code', 'County FIPS Code', 'Area Name', 'State Name']]
```

```python
county_df['Code'] = county_df['State FIPS Code'] + county_df['County FIPS Code']
county_df['Name'] = county_df['Area Name'] + ', ' + county_df['State Name']
county_df = county_df[['Code', 'Name']]
```

```python
county_df
```

```python
json_data = county_df.set_index('Code')['Name'].to_dict()
json_data
```

```python
with open('geocodes_codes.json', 'w') as json_file:
    json.dump(json_data, json_file)
```
